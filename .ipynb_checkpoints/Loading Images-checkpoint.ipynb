```python
import cv2
```

```python
from pathlib import Path
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
import torch
import torchvision.transforms as transforms
```

```python
caminhoImagem = Path('Anexos, Imagens e Videos/eye_image.jpg')
imagem = mpimg.imread(str(caminhoImagem))

# origin='upper' (Muda a legenda do eixo y com a origem no canto superior esquerdo)
plt.imshow(imagem, origin='upper')
plt.show()
```

```python
imagePath = Path('Anexos, Imagens e Videos/eye_image.jpg')
img = cv2.imread(str(imagePath))
```

```python
transform = transforms.ToTensor()
tensor = transform(img)
print(tensor.shape)
mean_img = tensor.mean((0))
mean_img.shape
```

```python
plt.imshow(mean_img, origin='upper',cmap='gray')
plt.show()
```
